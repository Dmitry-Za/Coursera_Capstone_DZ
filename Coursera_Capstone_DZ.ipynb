Read the dataframe using pandas read_html

```python
df=pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M") [0]
df
```

The format is not convenient. We will transform it into 3 separate lists of 1. postal code, 2. borough and 3. neighborhood (for the last one we will also replace "/" with ","), and on the next step we will combine them into one dataframe.

```python
i=0
j=0

list_all=[]
list_ind=[]
list_bor=[]
list_nei=[]

i=0

while (i<=19):
    j=0
    
    while (j<=8):
        list_all.append(df.iloc[i,j])
        list_ind.append(str(df.iloc[i,j])[0:3])
        list_bor.append(str(df.iloc[i,j])[3:str(df.iloc[i,j]).find("(")])
        list_nei.append(str(df.iloc[i,j])[str(df.iloc[i,j]).find("(")+1:str(df.iloc[i,j]).find(")")])
        j=j+1
    i=i+1

list_nei1 = [sub.replace('/', ',') for sub in list_nei] 
```

Here we put 3 lists together into 1 dataframe:

```python
#d = {'PostalCode': [list_ind], 'Borough': [list_bor],'Neighborhood':[list_nei]}
df = pd.DataFrame(list(zip(list_ind, list_bor,list_nei1)), 
               columns =['PostalCode', 'Borough','Neighborhood']) 
df 
```

Now we remove Boroughs that have no assigned values, and then remove the lines where 2 Boroughs are combined 

```python
df_filt1=df[df.Borough!="Not assigne"]
df_filt1
```

```python
df_filt1=df_filt1[df_filt1.Borough.str.len()<40]
df_filt1
```

And finally we reset the index

```python
df_filt1.reset_index(inplace = True, drop = True) 
df_filt1
```
